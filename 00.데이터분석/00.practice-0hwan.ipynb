```python
jumin = pd.read_csv('서울시주민등록인구(구별)통계.txt', sep='\t', skiprows=0)
jumin.head(3)
```

```python

jumin2 = jumin.set_index("자치구")
jumin2 = jumin2.drop(['세대당인구','인구.1','인구.2', '인구.4','인구.5','인구.7','인구.8',"세대","기간"], axis=1)
jumin2.rename(columns={"인구" : "전체 인구", "인구.3" : "한국인 합계", "인구.6" : "외국인 합계"},inplace=True)
jumin4 = jumin2.iloc[2:]
jumin4 = jumin4.iloc[1:]
jumin4.head(3)
```

```python
#데이터 값 정수로 변환
for column in jumin4.columns[:]:
    jumin4[column] = jumin4[column].apply(lambda x: int(x.replace(',','')))

jumin4.head(3)
```
